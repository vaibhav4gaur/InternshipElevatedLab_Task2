```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
Titanic = pd.read_csv('/Users/vaibhavgaurdeveloper/Downloads/Titanic-Dataset.csv')
```

```python
print("/n ---First 5 Rows---")
print(Titanic.head())
```

```python
print(f"\nDataset Shape: {Titanic.shape}")
```

```python
print("\n--- Dataset Info ---")
print(Titanic.info())
```

```python
print("\n--- Summary Statistics ---")
print(Titanic.describe(include='all'))
```

```python
print("\n--- Missing Values ---")
print(Titanic.isnull().sum())
```

```python
Titanic.hist(figsize=(12, 8), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numerical Features", fontsize=16)
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', data=Titanic)
plt.title("Boxplot of Age by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(Titanic.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Titanic Dataset")
plt.show()
```

```python
print(Titanic.columns.tolist())
```

```python
sns.pairplot(Titanic.dropna(), hue="Survived")
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
fig = px.scatter(Titanic, x="Age", y="Fare", color="Survived", title="Age vs Fare with Survival Status")
fig.show()
```

```python
print("\n--- Observations ---")
print("- Higher class passengers (Pclass=1) tend to be older.")
print("- Some outliers exist in Fare and Age.")
print("- Positive correlation observed between Fare and Survival.")
print("- Visuals show age and fare distribution varies by survival status.")
```

```python
print("\n--- Skewness ---")
print(Titanic.skew(numeric_only=True))
```
